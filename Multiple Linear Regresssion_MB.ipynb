# Preparing a Model

```python
import statsmodels.formula.api as smf
model1=smf.ols('MPG~WT+VOL+SP+HP',data=cars).fit()
model1.summary()
```

# Simple Linear Regression models

```python
ml_v=smf.ols('MPG~VOL',data=cars).fit()
ml_v.summary()
```

```python
ml_w=smf.ols('MPG~WT',data=cars).fit()
ml_w.summary()
```

```python
ml_vw=smf.ols('MPG~WT+VOL',data=cars).fit()
ml_vw.summary()
```

# Calculating VIF- Variance Inflation Factor

```python
rsq_hp=smf.ols('HP~WT+VOL+SP',data=cars).fit().rsquared
vif_hp=1/(1-rsq_hp)

rsq_wt= smf.ols('WT~HP+VOL+SP',data=cars).fit().rsquared
vif_wt= 1/(1-rsq_wt)

rsq_vol= smf.ols('VOL~HP+SP+WT',data=cars).fit().rsquared
vif_vol= 1/(1-rsq_vol)

rsq_sp =smf.ols('SP~HP+WT+VOL',data=cars).fit().rsquared
vif_sp= 1/(1-rsq_sp)

  # sorting vif values in a data frame
d1= {'Variables':['HPow','Wght','VOLum','Spd'],'VIF':[vif_hp,vif_wt,vif_vol,vif_sp]}
vif_frame= pd.DataFrame(d1)
vif_frame
```

# Residual Analysis

### Test for Normality of Residuals (Q-Q plot)

```python
res=model1.resid
res
```

```python
res.mean()
```

```python
import statsmodels.api as sm
qqplot=sm.qqplot(res,line='q')
plt.title('Test for Normality of Residuals(Q-Q plot)')
plt.show
```

```python
list(np.where(model1.resid>10))
```

# Residual Plot for Homoscedasticity

```python
def get_standardized_values (vals):
    return(vals-vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model1.fittedvalues),
           get_standardized_values(model1.resid))
plt.title('Residual Plot')
plt.xlabel('standardized fitted values')
plt.ylabel('standardized residual values')
plt.show()
```
